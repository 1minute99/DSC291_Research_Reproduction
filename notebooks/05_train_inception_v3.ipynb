# 05 — Train InceptionV3 on the spurious dataset (extension)
Proposal "New experiments": test whether MILAN's text-shortcut finding generalises beyond ResNet18 to a deeper, multi-branch architecture.

Settings come from `configs/inception_v3.yaml` and match the ResNet18 run (AdamW lr=1e-4, early-stop on val loss, patience=4) so the only change is the architecture. Two InceptionV3 specifics: **299×299** input and an **auxiliary classifier head** (folded into the training loss with weight 0.4).

**Expected outcome:** clean validation accuracy comparable to ResNet18 (~85-95%); we measure adversarial-test accuracy separately and expect a clear drop if the text shortcut took hold.

```python
import os, sys, yaml, torch
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
os.environ.setdefault('MILAN_DATA_DIR', str(Path.cwd().parent / 'data'))
os.environ.setdefault('MILAN_MODELS_DIR', str(Path.cwd().parent / 'models'))
from milan_repro.train.train_inception_v3 import train

cfg = yaml.safe_load(open('../configs/inception_v3.yaml'))
version_dir = Path(os.environ['MILAN_DATA_DIR']) / 'imagenet-spurious-text' / '50pct'
ckpt = Path(os.environ['MILAN_MODELS_DIR']) / 'inception_v3_spurious.pth'
# CUDA if present, else Apple GPU (mps) when available, else cpu.
device = 'cuda' if torch.cuda.is_available() else ('mps' if torch.backends.mps.is_available() else 'cpu')
print(f'Using device: {device}')
info = train(cfg, version_dir, ckpt, device=device)
info['best_val_loss']
```

```python
# Evaluate clean and adversarial test accuracy.
from torch.utils.data import DataLoader
from milan_repro.train.train_inception_v3 import load_trained
from milan_repro.data.spurious_dataset import load as load_spurious
from milan_repro.editing.evaluate import _accuracy

model = load_trained(ckpt, device=device)
# Inception wants 299x299 inputs, matching the training config.
_train_full, test_set = load_spurious(version_dir, image_size=cfg['data']['image_size'])
loader = DataLoader(test_set, batch_size=cfg['train']['batch_size'], num_workers=2)
adv_acc = _accuracy(model, loader, device)
print(f'adversarial-test acc: {adv_acc:.4f}')
```
